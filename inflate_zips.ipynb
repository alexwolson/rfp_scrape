```python
from pathlib import Path
from os import system
```

```python
root = Path('data')
for file in root.iterdir():
    if file.suffix == '.zip':
        html_name = Path(file.as_posix().replace('zip','html'))
        if html_name.exists():
            directory_name = file.parts[-1].split(".")[0]
            system(f'mkdir data/{directory_name}')
            system(f'mv {file.as_posix()} data/{directory_name}')
            system(f'mv {html_name.as_posix()} data/{directory_name}')
```

```python
for folder in root.iterdir():
    if folder.is_dir():
        for file in folder.iterdir():
            if file.suffix == '.zip':
                system(f'unzip {file.as_posix()} -d {folder.as_posix()}/')
```
